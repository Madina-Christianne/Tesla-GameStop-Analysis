```python
import yfinance as yf

# Télécharger les données boursières de Tesla (TSLA)
tesla = yf.Ticker("TSLA")

# Obtenir l'historique des prix
tesla_history = tesla.history(period="max")

# Afficher les 5 premières lignes
print(tesla_history.head())
import matplotlib.pyplot as plt

# Tracer l’évolution du prix de clôture
tesla_history['Close'].plot(figsize=(10,5), title="Évolution du prix de l'action Tesla")
plt.xlabel("Date")
plt.ylabel("Prix de clôture (USD)")
plt.show()

import yfinance as yf

# Télécharger les données de Tesla
tesla = yf.Ticker("TSLA")

# Obtenir les revenus
tesla_revenue = tesla.financials.loc["Total Revenue"]
print(tesla_revenue)


import pandas as pd
import matplotlib.pyplot as plt

# Transformer les revenus en DataFrame
tesla_revenue = tesla.financials.loc["Total Revenue"].reset_index()
tesla_revenue.columns = ["Date", "Revenue"]

# Supprimer les valeurs NaN
tesla_revenue.dropna(inplace=True)

# Convertir la colonne Revenue en milliards
tesla_revenue["Revenue"] = tesla_revenue["Revenue"] / 1e9  # Convertir en milliards de dollars

print(tesla_revenue)

# Tracer les revenus de Tesla
plt.figure(figsize=(10,5))
plt.plot(tesla_revenue["Date"], tesla_revenue["Revenue"], marker='o', linestyle='-', color='b')

# Ajouter les titres et labels
plt.title("Évolution des revenus de Tesla")
plt.xlabel("Année")
plt.ylabel("Revenu (Milliards USD)")
plt.grid(True)

# Afficher le graphique
plt.show()

# Télécharger les données de GameStop (GME)
gamestop = yf.Ticker("GME")

# Obtenir l'historique des prix
gme_history = gamestop.history(period="max")

# Afficher les 5 premières lignes
print(gme_history.head())

# Tracer l’évolution du prix de clôture
gme_history['Close'].plot(figsize=(10,5), title="Évolution du prix de l'action GameStop")
plt.xlabel("Date")
plt.ylabel("Prix de clôture (USD)")
plt.show()
# Obtenir les revenus de GameStop
gme_revenue = gamestop.financials.loc["Total Revenue"].reset_index()
gme_revenue.columns = ["Date", "Revenue"]

# Supprimer les valeurs NaN
gme_revenue.dropna(inplace=True)

# Convertir la colonne Revenue en milliards
gme_revenue["Revenue"] = gme_revenue["Revenue"] / 1e9  # Convertir en milliards de dollars

# Afficher les revenus
print(gme_revenue)

# Tracer les revenus de GameStop
plt.figure(figsize=(10,5))
plt.plot(gme_revenue["Date"], gme_revenue["Revenue"], marker='o', linestyle='-', color='r')

# Ajouter les titres et labels
plt.title("Évolution des revenus de GameStop")
plt.xlabel("Année")
plt.ylabel("Revenu (Milliards USD)")
plt.grid(True)

# Afficher le graphique
plt.show()

# Tracer les revenus de Tesla et GameStop sur le même graphique
plt.figure(figsize=(10,5))

# Tracer Tesla (en bleu)
plt.plot(tesla_revenue["Date"], tesla_revenue["Revenue"], marker='o', linestyle='-', color='b', label="Tesla")

# Tracer GameStop (en rouge)
plt.plot(gme_revenue["Date"], gme_revenue["Revenue"], marker='o', linestyle='-', color='r', label="GameStop")

# Ajouter les titres et labels
plt.title("Comparaison des revenus de Tesla et GameStop")
plt.xlabel("Année")
plt.ylabel("Revenu (Milliards USD)")
plt.legend()
plt.grid(True)

# Afficher le graphique
plt.show()
```
